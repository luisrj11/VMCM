```python
# Monte Carlos Np-3D
# Common imports
import os

# Where to save the figures and data files
PROJECT_ROOT_DIR = "Results-Np-3D"
FIGURE_ID = "Results-Np-3D/FigureFiles"
DATA_ID = "Results-Np-3D/VMCHarmonic"

if not os.path.exists(PROJECT_ROOT_DIR):
    os.mkdir(PROJECT_ROOT_DIR)

if not os.path.exists(FIGURE_ID):
    os.makedirs(FIGURE_ID)

if not os.path.exists(DATA_ID):
    os.makedirs(DATA_ID)

def image_path(fig_id):
    return os.path.join(FIGURE_ID, fig_id)

def data_path(dat_id):
    return os.path.join(DATA_ID, dat_id)

def save_fig(fig_id):
    plt.savefig(image_path(fig_id) + ".png", format='png')

outfile = open(data_path("VMCHarmonic.dat"),'w')
```

```python
%matplotlib inline

# VMC for the one-dimensional harmonic oscillator
# Brute force Metropolis, no importance sampling and no energy minimization
from math import exp, sqrt
from random import random, seed
import numpy as np
import matplotlib.pyplot as plt
from decimal import *
# Trial wave function for the Harmonic oscillator in 3D
def WaveFunction(r,alpha,NumberParticles,Dimension):
    r_sum = 0 
    for i in range(NumberParticles):
        for j in range(Dimension):
            r_sum += r[i][j]*r[i][j]
    return exp(-alpha*r_sum)

# Local energy  for the Harmonic oscillator for Np 3D in 3D
def LocalEnergy(r,alpha,NumberParticles,Dimension):
    r_sum = 0 
    for i in range(NumberParticles):
        for j in range(Dimension):
            r_sum += r[i][j]*r[i][j]
    return NumberParticles*Dimension*alpha + (0.5 - 2*alpha*alpha)*r_sum
```

```python
# The Monte Carlo sampling with the Metropolis algo
# The jit decorator tells Numba to compile this function.
# The argument types will be inferred by Numba when the function is called.

def MonteCarloSampling(
    NumberParticles = 5,
    Dimension = 1,
    # Number Monte Carlos cicles
    NumberMCcycles= 10**5,
    StepSize = 1.0,
    # Number varations of Alpha 
    VariationsAlfa = 20,
    AlphaStart= 0.1,
    StespAlpha = .05):
    #----------------------------------------------------------#    
    PositionOld = np.zeros((NumberParticles,Dimension), np.double)
    PositionNew = np.zeros((NumberParticles,Dimension), np.double)
    #----------------------------------------------------------#
    # Save all variations  
    Energies = np.zeros(VariationsAlfa)
    ExactEnergies = np.zeros(VariationsAlfa)
    AlphaValues = np.zeros(VariationsAlfa)
    #----------------------------------------------------------#
    # seed starts random numbers  
    seed()
    #----------------------------------------------------------#
    # Start variational parameter
    alpha = AlphaStart
    for ia in range(MaxVariations):
        alpha += StespAlpha
        AlphaValues[ia] = alpha
        energy = 0.0
        energy2 = 0.0
        #----------------------------------------------------------#
        # Initial position
        for i in range(NumberParticles):
            for j in range(Dimension):
                PositionOld[i][j] = StepSize * (random() - .5)
        wfold = WaveFunction(PositionOld,alpha,NumberParticles,Dimension)
        #----------------------------------------------------------#
        # Loop over Monte Carlos cicles (MCcycles)
        for MCcycle in range(NumberMCcycles):
            #----------------------------------------------------------#
            #Trial position moving one particle at the time
            for i in range(NumberParticles):
                for j in range(Dimension):
                    PositionNew[i][j] = PositionOld[i][j] + StepSize * (random() - .5)
                wfnew = WaveFunction(PositionNew,alpha,NumberParticles,Dimension)
                #----------------------------------------------------------#
                #Metropolis test to see whether we accept 
                if random() < wfnew**2 / wfold**2:
                    for j in range(Dimension):
                        PositionOld[i][j] = PositionNew[i][j]
                    wfold = wfnew
                #----------------------------------------------------------#
            DeltaE = LocalEnergy(PositionOld,alpha,NumberParticles,Dimension)
            energy += DeltaE
            energy2 += DeltaE**2
            #----------------------------------------------------------#
        #----------------------------------------------------------#
        # We calculate mean, variance and error ...
        energy /= NumberMCcycles
        energy2 /= NumberMCcycles
        variance = energy2 - energy**2
        error = sqrt(variance/NumberMCcycles)
        #----------------------------------------------------------#
        # Saving each iterations
        Energies[ia] = energy 
        Variances[ia] = variance 
        #------------ ---------------------------------------#
        # Writing in a external file("VMCHarmonic.dat") for each iterations
        # the path it is renamed with the alias (outfile)
        outfile.write('%f %f %f %f \n' %(alpha,energy,variance,error))
        #----------------------------------------------------------# 
    return Energies, AlphaValues, Variances
```

```python
#----------------------------------------------------------#
#Here starts the main program with variable declarations
N=3  #Number of particles
D=3  #Dimension: D=1,2,3
MaxVariations = 20
Energies = np.zeros((MaxVariations))
#ExactEnergies = np.zeros((MaxVariations))
#ExactVariance = np.zeros((MaxVariations))
Variances = np.zeros((MaxVariations))
AlphaValues = np.zeros(MaxVariations)
#----------------------------------------------------------#
(Energies, AlphaValues, Variances) = MonteCarloSampling()
outfile.close()
ExactEnergies = (0.5 - 2*AlphaValues*AlphaValues)*((D*N)/(4*AlphaValues)) + D*N*AlphaValues 
ExactVariance = 0.0625*N*D*((D+2)/(AlphaValues*AlphaValues))*(0.5 - 2*AlphaValues*AlphaValues)**2 + 0.5*N*(0.5 - 2*AlphaValues*AlphaValues)*D**2 + N*(D*AlphaValues)**2  - ExactEnergies*ExactEnergies
#----------------------------------------------------------#
#simple subplot
plt.subplot(2, 1, 1)
plt.plot(AlphaValues, Energies, 'o-',AlphaValues, ExactEnergies,'r-')
plt.title('Energy and variance')
plt.ylabel('Dimensionless energy')
plt.subplot(2, 1, 2)
plt.plot(AlphaValues, Variances, '.-',AlphaValues, ExactVariance,'r-')
plt.xlabel(r'$\alpha$', fontsize=15)
plt.ylabel('Variance')
save_fig("VMCHarmonic")
plt.show()
#----------------------------------------------------------#
#nice printout with Pandas
import pandas as pd
from pandas import DataFrame
#data ={'Alpha':AlphaValues, 'Energy':Energies,'Exact Energy':ExactEnergies,'Variance':Variances,'Exact Variance':ExactVariance,}
data ={'Alpha':AlphaValues, 'Energy':Energies,'Variance':Variances,}
frame = pd.DataFrame(data)
print(frame)
```
