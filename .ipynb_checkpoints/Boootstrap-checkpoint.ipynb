```python
'''
=======================================================================|
VMC for the 3D-dimensional and N particles using Bootstrap technique
Consideratios:
    1 Using Harmoninc oscilator as wave funtion
      - We suposse to know the analitically expretion
        of the local energy
    2 Brute force Metropolis(No importance sampling)
    3 No energy minimization(Varying alpha value)
'''

#---------------------------------------------------------------------#
# Pakages
#---------------------------------------------------------------------#
%matplotlib inline
from math import exp, sqrt
from random import random, seed
import numpy as np
import matplotlib.pyplot as plt
from decimal import *
import time
#---------------------------------------------------------------------#

#---------------------------------------------------------------------#
# Trial wave function for the Harmonic oscillator in 3D
#---------------------------------------------------------------------#
def WaveFunction(r,alpha,NumberParticles,Dimension):
    r_sum = 0 
    for i in range(NumberParticles):
        for j in range(Dimension):
            r_sum += r[i][j]*r[i][j]
    return exp(-alpha*r_sum)
#---------------------------------------------------------------------#
# Local energy  for the Harmonic oscillator for Np in 3D
#---------------------------------------------------------------------#
def LocalEnergy(r,alpha,NumberParticles,Dimension):
    r_sum = 0 
    for i in range(NumberParticles):
        for j in range(Dimension):
            r_sum += r[i][j]*r[i][j]
    return NumberParticles*Dimension*alpha + (0.5 - 2*alpha*alpha)*r_sum
    
#---------------------------------------------------------------------#
```

```python
import numpy as np
from numpy.random import randint, randn
from time import time
from scipy.stats import norm
import matplotlib.pyplot as plt
from random import random, seed

#---------------------------------------------------------------------#
# Returns mean of bootstrap samples       
def Mean(Data): # To calculate the mean 
    '''Here can be used numpy (np.mean(Data))'''
    NumberData = len(Data)
    Averange = 0
    for iData in range(NumberData):
        Averange += Data[iData]
    Averange = float(Averange)
    LengthData = float(NumberData)
    Averange /= NumberData
    return Averange
#---------------------------------------------------------------------#

# Returns varaince of bootstrap samples   
def Variance(Data): # To calculate the variance  
    '''Here can be used the numpy (np.std(Data))'''
    NumberData = len(Data)
    mean = Mean(Data)
    variance = 0
    for iData in range(NumberData):
        variance += (Data[iData] - mean)**2
    variance /= NumberData
    return variance
#---------------------------------------------------------------------#

# Creat a date base
def CreatDataBase(NDataPoints,     # Number the data point   
                  NumberParticles, # Number particle
                  Dimension,       # Dimension
                  StepSize =1):    # Step size normaly is equal to 1
    SaveData = []  # To save the data
    # To make the data base
    for  iDate in range(NDataPoints): 
        Positions = np.zeros((NumberParticles,Dimension), np.double)
        # Moving one particle at the time
        for i in range(NumberParticles):
            for j in range(Dimension):
                Positions[i][j] = StepSize * (random() - .5)
        SaveData.append(Positions)
    return  SaveData
#---------------------------------------------------------------------#     
```

```python
# Bootstrap algorithm  

def MonteCarloBootstrap(
    DataBase,            
    NumberParticles ,
    Dimension ,
    NumberResampling,    # Number of resampling 
    NumberMCcycles,      # Number Monte Carlos cicles
    StepSize = 1.0,
    VariationsAlfa = 20, # Number varations of Alpha 
    AlphaStart= 0.1,     # Where alpha start
    StespAlpha = .05):   # Alpha step
    #----------------------------------------------------------#
    # Number of the data base
    #----------------------------------------------------------#
    Ndate = len(DataBase)
    #----------------------------------------------------------#    
    PositionOld = np.zeros((NumberParticles,Dimension), np.double)
    PositionNew = np.zeros((NumberParticles,Dimension), np.double)
    #----------------------------------------------------------#
    # Save all variations  
    #----------------------------------------------------------#
    AverangeEnergies = np.zeros(VariationsAlfa)
    AverangeVariances= np.zeros(VariationsAlfa)
    #----------------------------------------------------------#
    Energies = np.zeros((NumberResampling,VariationsAlfa),np.double)
    AlphaValues = np.zeros(VariationsAlfa)
    Variances = np.zeros((NumberResampling,VariationsAlfa),np.double)
    #----------------------------------------------------------#
    # Start variational parameter alpha
    #----------------------------------------------------------#
    alpha = AlphaStart
    for ia in range(MaxVariations):
        alpha += StespAlpha
        AlphaValues[ia] = alpha
        #----------------------------------------------------------#
        # Number of resampling
        #----------------------------------------------------------#
        for Nrs in range (NumberResampling):
            energy = 0.0
            energy2 = 0.0
            #----------------------------------------------------------#
            # Initial position
            #----------------------------------------------------------#
            for i in range(NumberParticles):
                for j in range(Dimension):
                    PositionOld[i][j] = DataBase[randint(0,Ndate)][i][j]
            wfold = WaveFunction(PositionOld,alpha,NumberParticles,Dimension)
            #----------------------------------------------------------#
            # Loop over Monte Carlos cicles (MCcycles)
            #----------------------------------------------------------#
            for MCcycle in range(NumberMCcycles):
                #----------------------------------------------------------#
                # Trial position moving one particle at the time
                #----------------------------------------------------------#
                for i in range(NumberParticles):
                    for j in range(Dimension):
                        PositionNew[i][j] = PositionOld[i][j] + DataBase[randint(0,Ndate)][i][j]
                    wfnew = WaveFunction(PositionNew,alpha,NumberParticles,Dimension)
                    #----------------------------------------------------------#
                    # Metropolis test to see whether we accept 
                    if random() < wfnew**2 / wfold**2:
                        for j in range(Dimension):
                            PositionOld[i][j] = PositionNew[i][j]
                        wfold = wfnew
                    #----------------------------------------------------------#
                DeltaE = LocalEnergy(PositionOld,alpha,NumberParticles,Dimension)
                energy += DeltaE
                energy2 += DeltaE**2
                #----------------------------------------------------------#
            #----------------------------------------------------------#
            # Calculate mean, variance and error for Monte Carlos cycles
            #----------------------------------------------------------#
            energy /= NumberMCcycles
            energy2 /= NumberMCcycles
            Variance = energy2 - energy**2
            #error = sqrt(variance/NumberMCcycles)
            #----------------------------------------------------------#
            # Saving iterations for each resampling
            #----------------------------------------------------------#
            Energies[Nrs][ia] = energy 
            Variances[Nrs][ia] = Variance 
            #------------ ---------------------------------------#
            # Writing in a external file("VMCHarmonic.dat") for each iterations
            # the path it is renamed with the alias (outfile)
            #outfile.write('%f %f %f %f \n' %(alpha,energy,variance,error))
            #----------------------------------------------------------# 
    #----------------------------------------------------------#
    # Calculate averange and variance of resapling
    #----------------------------------------------------------#
    for ia in range(VariationsAlfa):
        AverangeEnergies[ia] = Mean(Energies[:,ia])
        AverangeVariances[ia] = Mean(Variances[:,ia])
    #----------------------------------------------------------#
    # end 
    #----------------------------------------------------------#
    return AverangeEnergies, AlphaValues,  AverangeVariances
```

```python
import time
inicio = time.time()

#----------------------------------------------------------#
# Time CPU
inicio = time.time()
#----------------------------------------------------------#
#Here starts the main program with variable declarations

DataPoints = 10**5        # Number of data base

NumberParticles = 5       # Number of particles
Dimension = 2             # Dimension: D=1,2,3
NumberResampling = 10**1  # Number of resampling
NumberMCcycles = 10**3    # Number Monte Carlos cycles
VariationsAlfa = 20       # Number variations of alpha

DataBase = CreatDataBase(DataPoints,NumberParticles,Dimension,StepSize =1)
#----------------------------------------------------------#

N = NumberParticles  #Number of particles
D = Dimension       #Dimension: D=1,2,3
MaxVariations = VariationsAlfa
Energies = np.zeros((MaxVariations))
ExactEnergies = np.zeros((MaxVariations))
ExactVariance = np.zeros((MaxVariations))
Variances = np.zeros((MaxVariations))
AlphaValues = np.zeros(MaxVariations)
#----------------------------------------------------------#
(Energies, AlphaValues, Variances) = MonteCarloBootstrap(
    DataBase,
    NumberParticles ,
    Dimension ,
    NumberResampling ,     # NUmber resampling   
    NumberMCcycles,       # Number Monte Carlos cicles
    StepSize = 1.0, 
    VariationsAlfa = 20,  # Number varations of Alpha
    AlphaStart= 0.1,
    StespAlpha = .05)
#outfile.close()
ExactEnergies = (0.5 - 2*AlphaValues*AlphaValues)*((D*N)/(4*AlphaValues)) + D*N*AlphaValues 
#ExactVariance = ((D*N)**2)*(1/4)*(1 +((1 - 4*AlphaValues**2)**2)*(3/((4*AlphaValues)**2))) - ExactEnergies*ExactEnergies 
ExactVariance = 0.0625*N*D*((D+2)/(AlphaValues*AlphaValues))*(0.5 - 2*AlphaValues*AlphaValues)**2 + 0.5*N*(0.5 - 2*AlphaValues*AlphaValues)*D**2 + N*(D*AlphaValues)**2  - ExactEnergies*ExactEnergies
#----------------------------------------------------------#
#simple subplot
plt.subplot(2, 1, 1)
plt.plot(AlphaValues, Energies, 'o-',AlphaValues, ExactEnergies,'r-')
plt.title('Energy and variance')
plt.ylabel('Dimensionless energy')
plt.xlabel(r'$\alpha$', fontsize=15)

#plt.subplot(2, 1, 2)
#plt.plot(AlphaValues, Variances, '.-',AlphaValues, ExactVariance,'r-')
#plt.ylabel('Variance')
#plt.xlabel(r'$\alpha$', fontsize=15)
##save_fig("VMCHarmonic")
#plt.show()

#----------------------------------------------------------#
#nice printout with Pandas
import pandas as pd
from pandas import DataFrame
data ={'Alpha':AlphaValues, 'Energy':Energies,'Exact Energy':ExactEnergies,'Variance':Variances,'Exact Variance':ExactVariance,}
#data ={'Alpha':AlphaValues, 'Energy':Energies,'Variance':Variances,}
frame = pd.DataFrame(data)
print(frame)

#----------------------------------------------------------#

fin = time.time()
print("---------------------------------------------------------------")
print('CPU time consuming')
print("---------------------------------------------------------------")
print("CPU time =",fin-inicio, "seconds")
print("CPU time =",(fin-inicio)/60, "minutes")
print("CPU time =",(fin-inicio)/3600, "hours")
print("---------------------------------------------------------------")
print('Number resampling ',NumberResampling)
print("---------------------------------------------------------------")
print('Number Monte Carlos cycles ',NumberMCcycles)
print("---------------------------------------------------------------")
```

```python
from numpy import *
from numpy.random import randint, randn
from time import time
from scipy.stats import norm
import matplotlib.pyplot as plt

# Returns mean of bootstrap samples                                                                                                                                                
def stat(data):
    return mean(data)

# Bootstrap algorithm                                                                                                                                                              
def bootstrap(data, statistic, R):
    t = zeros(R); n = len(data); inds = arange(n); t0 = time()

    # non-parametric bootstrap                                                                                                                                                     
    for i in range(R):
        t[i] = statistic(data[randint(0,n,n)])

    # analysis                                                                                                                                                                     
    print("Runtime: %g sec" % (time()-t0)); print("Bootstrap Statistics :")
    print("original           bias      std. error")
    print("%8g %8g %14g %15g" % (statistic(data), std(data),\
                             mean(t), \
                             std(t)))
    return t


mu, sigma = 100, 15
datapoints = 10000
x = mu + sigma*random.randn(datapoints)
# bootstrap returns the data sample                                                                                                          t = bootstrap(x, stat, datapoints)
# the histogram of the bootstrapped  data  
t = bootstrap(x, stat, datapoints)
# the histogram of the bootstrapped  data                                            
n, binsboot, patches = plt.hist(t, bins=50, density='true',histtype='bar', color='red', alpha=0.75)

# add a 'best fit' line                                                                                                                                                          
y = norm.pdf( binsboot, mean(t), std(t))
lt = plt.plot(binsboot, y, 'r--', linewidth=1)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.axis([99.5, 100.6, 0, 3.0])
plt.grid(True)

plt.show()
     
```
